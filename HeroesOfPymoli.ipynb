```python
# 1. Dependencies and Setup

import pandas as pd

# File to Load 
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data["Total Price"] = purchase_data["Price"]*1
is_Male =  purchase_data['Gender']=='Male'
purchase_data_male = purchase_data[is_Male]
is_Female =  purchase_data['Gender']=='Female'
purchase_data_female = purchase_data[is_Female]
is_Other =  purchase_data['Gender']=='Other / Non-Disclosed'
purchase_data_other = purchase_data[is_Other]
gender_count = purchase_data["Gender"].value_counts()

purchase_data_topspend = purchase_data.groupby(['SN'])
topspender_count_df = purchase_data["SN"].value_counts()
top_item = purchase_data["Item Name"].value_counts()
Valuable_item = purchase_data.sort_values(['Price','SN'],ascending=False).groupby('Item ID').head(3)
```

```python
# 2. Player Count
df_players = pd.DataFrame(
[[purchase_data["SN"].nunique()]],
index=[0],
columns=['Total Players'])
df_players
```

```python
# 3. Purchasing Analysis (Total)
df_total = pd.DataFrame(
[[purchase_data["Item Name"].nunique(), purchase_data["Price"].mean(), purchase_data["Purchase ID"].nunique(), purchase_data["Price"].sum()]],
index=[0],
columns=['No of Unique Items', 'Average Price', 'Number of Purchase', 'Total Revenue'])
df_total.round(2)
```

```python
# 4. Gender Demographics
#gender_count = purchase_data["Gender"].value_counts()
df_demo = pd.DataFrame(
[[purchase_data_male["SN"].nunique(),gender_count [0]/len(purchase_data["Gender"])*100],
[purchase_data_female["SN"].nunique(),gender_count [1]/len(purchase_data["Gender"])*100],
[purchase_data_other["SN"].nunique(),gender_count [2]/len(purchase_data["Gender"])*100]],
index=['Male','Female','Other / Non-Disclosed'],
columns=['Total Count', 'Percentage of Players'])
df_demo.round(2)
```

```python
# 5. Purchasing Analysis (Gender)
df_PAG = pd.DataFrame(
[['','','',''],
[gender_count [1],purchase_data_female["Price"].mean(),purchase_data_female["Price"].sum(),purchase_data_female["Price"].sum()/purchase_data_female["SN"].nunique()],
[gender_count [0],purchase_data_male["Price"].mean(),purchase_data_male["Price"].sum(),purchase_data_male["Price"].sum()/purchase_data_male["SN"].nunique()],
[gender_count [2],purchase_data_other["Price"].mean(),purchase_data_other["Price"].sum(),purchase_data_other["Price"].sum()/purchase_data_other["SN"].nunique()]], 
index=['Gender','Female','Male','Other / Non-Disclosed'],
columns=['Purchase Count', 'Average Purchase Price','Total Purchase Value','Avg Total Purchase per Person'])

df_PAG.round(2)
```

```python
# 6. Age Demographics

# Creating a group based off of the bins
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
# Create the names for bins
group_names = ["<10","10-14","15-19","20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data["Summary"] = pd.cut(purchase_data["Age"], bins, labels=group_names, include_lowest=True)
purchase_data= purchase_data.groupby("Summary")
SNN = purchase_data["SN"].unique()
#purchase_data.count().head(10)
```
